```python
import numpy as np
import matplotlib.pyplot as plt


# Step 1: Read the file
labels = []
values = []

data = np.genfromtxt('average_energy.txt', delimiter='\t', dtype=str)
labels = data[:,0] # First column (stages)
values = data[:,1].astype(float) # Second column (averaged data)

plt.bar(labels, values)
plt.xlabel('Values')
plt.title('Stage Values')

plt.show()
```
